```python
X = df.drop(columns = ['Outcome'])
y = df['Outcome']
```

```python
from sklearn.model_selection import train_test_split
train_x,test_x,train_y,test_y = train_test_split(X,y,test_size = 0.2,random_state = 35)
```

```python
from sklearn.preprocessing import StandardScaler
standard = StandardScaler()
```

```python
train_standart_x = standard.fit_transform(train_x)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=13)
```

```python
knn.fit(train_standart_x,train_y)
```

```python
train_standart_y = standard.fit_transform(test_x)
```

```python
predict_test = knn.predict(train_standart_y)
```

```python
from sklearn.metrics import precision_score, recall_score, f1_score, accuracy_score
prec = precision_score(test_y,predict_test)
recall = recall_score(test_y,predict_test)
f1 = f1_score(test_y,predict_test)
accuracy = accuracy_score(test_y,predict_test)
print(prec)
print(recall)
print(f1)
print(accuracy)
```

```python
f1 = []
for i in range(1,60):
  knn = KNeighborsClassifier(n_neighbors=i)
  knn.fit(train_standart_x,train_y)
  predict_test = knn.predict(train_standart_y)
  f1_c = f1_score(test_y,predict_test)
  f1.append(f1_c)
```

```python
plt.figure(figsize=(22,6))
plt.plot(range(1,60),f1)
plt.xticks(range(1,60))
plt.grid()
plt.show()
```

```python
from sklearn.model_selection import cross_val_predict
predict = cross_val_predict(estimator = knn, X = X, y = y, cv = 5)

# Modelni baholash
from sklearn.metrics import classification_report
print("Classification Report: \n",classification_report(y, predict))
```

```python
from sklearn.model_selection import GridSearchCV
parameters = {'kernel':('linear', 'rbf'), 'C':[1, 10]}
clf = GridSearchCV(knn, parameters)
clf.fit(train_standart_x,train_y)
```
